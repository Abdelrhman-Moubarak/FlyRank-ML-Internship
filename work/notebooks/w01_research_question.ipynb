## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import json

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

stale_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)]
print(f"{len(stale_visible)} of {len(df)} pages are stale and still visible")

res = json.load(open("../../outputs/model_results.json"))
base = res["baseline"]["baseline_precision_at_50"]
rf   = res["models"]["random_forest"]["precision_at_50"]
print(f"Precision@50 - hand rule: {base:.3f}   model: {rf:.3f}")


age_by_trend = df.groupby("trend_direction")["content_age_days"].median()
print("\n")
print(age_by_trend.round(0).to_string())
```
